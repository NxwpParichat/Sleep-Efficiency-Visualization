#### **ภายใน Dataset มีข้อมูลดังต่อไปนี้**
  - ID = ตัวระบุที่ไม่ซ้ำกันสำหรับหัวข้อทดสอบแต่ละหัวข้อ
  - Age = อายุของผู้เข้ารับการทดสอบ
  - Gender = male or female
  - Bedtime = เวลาที่ผู้ทดสอบเข้านอนในแต่ละคืน
  - Wakeup time = เวลาที่ผู้ถูกทดสอบตื่นนอนทุกเช้า
  - Sleep duration = ระยะเวลาทั้งหมดที่ผู้ทดสอบนอนหลับ (เป็นชั่วโมง)
  - REM sleep percentage = เปอร์เซ็นต์ของเวลาการนอนหลับทั้งหมดที่ใช้ในช่วงระยะหลับฝัน วงจรนี้จะเกิดขึ้น 90 นาทีหลังเราหลับลึก
  - Deep sleep percentage = เปอร์เซ็นต์ของเวลาการนอนหลับทั้งหมดที่ใช้ในการนอนหลับลึก
  - Light sleep percentage = เปอร์เซ็นต์ของเวลานอนหลับทั้งหมดที่ใช้ในการนอนหลับตื้น
  - Awakenings = จำนวนครั้งที่ผู้ทดสอบตื่นขึ้นมาในตอนกลางคืน
  - Caffeine consumption = ปริมาณคาเฟอีนที่บริโภคใน 24 ชั่วโมงก่อนเข้านอน (เป็นมิลลิกรัม)
  - Alcohol consumption = ปริมาณแอลกอฮอล์ที่บริโภคใน 24 ชั่วโมงก่อนเข้านอน (หน่วยเป็นออนซ์)
  - Smoking status = ผู้ทดสอบสูบบุหรี่หรือไม่
  - Exercise frequency = จำนวนครั้งที่ผู้ทดสอบออกกำลังกายในแต่ละสัปดาห์


#### **ตัวแปรเป้าหมาย คือ**

# Machine Learning

```python
df = pd.read_csv('Sleep_Efficiency.csv')
df.dropna(inplace=True)

att = df[['Alcohol consumption']]
label = df['Sleep efficiency']

model = LinearRegression()
model.fit(att, label)

# ทำนายผลลัพธ์
ans = model.predict(att)

# พล็อตแผนภาพ
plt.figure(figsize=(12, 6))
plt.bar(df['Alcohol consumption'], df['Sleep efficiency'], color="darkcyan") 

# เพิ่มเส้นกราฟ
plt.plot(att, ans, color='r', linestyle='-', linewidth=3) 

plt.xlabel('Alcohol consumption (oz.)', fontsize=14)  
plt.ylabel('Sleep efficiency (%)', fontsize=14)  
plt.title('Alcohol Consumption vs. Sleep Efficiency', fontsize=16) 

# ประเมินค่าความแม่นยำของโมเดล
score = model.score(att, label)
print("Model Score:", score)
```

# หลังนำเสนอและลองดู Model score ใหม่ค่ะ🙏

```python
from sklearn.preprocessing import PolynomialFeatures

df = pd.read_csv('Sleep_Efficiency.csv')
df.dropna(inplace=True)

att = df[['Alcohol consumption', 'Exercise frequency', 'Age', 'Deep sleep percentage']]
label = df['Sleep efficiency']

pf = PolynomialFeatures(degree = 2) 
pf.fit(att)
att_new = pf.transform(att)

model = LinearRegression()
model.fit(att_new, label)

ans = model.predict(att_new)

score = model.score(att_new, label)
print("Model Score:", score)
```
